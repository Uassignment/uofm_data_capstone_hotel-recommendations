```python
import pandas as pd
import numpy as np
```

```python
url= "https://ralston-hotel-data-bucket.s3.amazonaws.com/Hotel_Reviews.csv"
```

```python
hotel_df = pd.read_csv(url)
```

```python
hotel_df.head()
```

```python
hotel_df.info()
```

```python
hotel_df.describe()
```

```python
#change review date from object to date/time
hotel_df["Review_Date"]=pd.to_datetime(hotel_df.Review_Date)
hotel_df.info()
```

```python
# % original rows remaining if null lat/long removed
512470/515738
```

```python
#drop null lat/long
hotel_df.dropna(inplace = True)
```

```python
hotel_df.info()
```

```python
# Replacing "United Kingdom with "UK"
hotel_df.Hotel_Address = hotel_df.Hotel_Address.str.replace("United Kingdom", "UK")
```

```python
#split country from address
hotel_df["Hotel_Country"] = hotel_df.Hotel_Address.apply(lambda x: x.split(' ')[-1])
print(hotel_df.Hotel_Country.unique())
```

```python
hotel_df.head()
```

```python
#add binning for reviewer score
bin_labels = ['Negative', 'Positive']
cut_bins = [0, 5, 10]
hotel_df['Sentiment'] = pd.cut(hotel_df['Reviewer_Score'], bins=cut_bins, labels=bin_labels)
```

```python
hotel_df.head()
```

```python
#Replace no negative and no positive review with null
hotel_df.replace({'No Negative': None, 'No Positive': None})
```
